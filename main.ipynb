`Please briefly outline the scenario`

A necklace company is meticulously preparing unique products for the upcoming New Year. It is to manufacture a necklace with the appropriate zodiac shape related to the customer's birthday. 
The signs of the zodiac related to birthdays are expected to be popular because they can express a person's personality and characteristics. In addition, to increase the uniqueness of the necklace, the company will combine diamonds where the brightest star is located among the major stars that make up the signs of the zodiacs. Thus, they have to find and mark the brightest star among the stars that make up the signs of the zodiac.  
To create the design of this product, try to develop a system to produce a draft figure out the shape of the zodiac and the brightest star composed of that signs of the zodiac. First, develop a program to get input on a customer's birthday to identify the shape of a distinct zodiac sign corresponding to that birthday, and to find the brightest star among them.

---



`Aim of project and the problem it is solving`

The project comprises four primary objectives. First, identify the Zodiac sign associated with your birthdate. Second, the task is to calculate the average coordinates of the major stars corresponding to the Zodiac to get an image from SDSS by using center coordinates. Following this, obtaining an image from the SDSS with the center coordinates. Third, mark the coordinates and connect them to the major stars using lines. Finally, the major stars need to be analyzed to identify the brightest one, and this one should be marked on the image.

---


`Overview of how code works`

1. User Input and Validation:
The program prompts the user to enter their birthday.
Validates the input to ensure it conforms to a valid date format with datetime library.

2. Determine Zodiac Sign:
Uses the provided birthday to determine the corresponding zodiac sign by retrieving this information from the 'zodiac_birth.json' file.

3. Calculate Zodiac Center:
Maps the zodiac sign to 'zodiac_stars.json' to obtain the coordinates of major stars that define the zodiac.Calculates the center Ra(Right Ascension), Dec(Declination) based on the mean of these star coordinates.

4. Retrieve Zodiac Image:  
Uses the calculated Ra, Dec to search for the Zodiac image to be used for visualization. in the SDSS (Sloan Digital Sky Survey).

5. Graph Mapping and Visualization:  
Maps the graph information in 'zodiac_stars.json' onto the retrieved zodiac and image.
Utilizes the BFS algorithm to navigate through each node in the graph. BFS is suitable for solving this problem.The low depth of graph degrees and the necessity to visit all nodes is the nature of the Zodiac.  This algorithm efficiently traverses through the graph, exploring nodes at the current depth before moving on to nodes at the next depth. It's a suitable approach for zodiac where the goal is to visit all nodes and represent their connectivity.

6. Highlighting Representation for the brightest star:  
The node with the brightest star is then marked with a distinct color, allowing for clear visualization. In this process, due to the unavailability of flux values for the major stars in the zodiac zodiac from SDSS, an alternative approach is taken. The algorithm retrieves images of the zodiac from the DSS (Digital Sky Survey) and compares pixel values to identify the brightest star. While flux values would be more appropriate for comparison, the absence of this data prompts the use of pixel intensity as a substitute. This method effectively determines the brightest star within the zodiac for visualization purposes.

* Some images in the SDSS used in this project are outside the boundaries of the SDSS dataset, so dark images may appear.

## Main Code

**`class Zodiac`**  

- `def get_zodiac_name(month, day)`  : Mapping the month and day to 'zodiiac_births.json' to get the appropriate zodiac name for that date.
<br>  

- `get_zodiac_centre(zodiac name)`   : calculates the mean of that star's coordinates(Ra,Dec), and it is used as a centre coordinates for the Zodiac.
<br>

- `get_brightest_star(Key stars coordinates info)`  : Image information of the major stars that consist of the zodiac is retrieved from DSS. Find the brightest star based on the brightest value among the pixels in the image. (Compare the pixel values in an alternative way because SDSS and DSS do not provide flux information about the stars that make up the zodiac.)


```python
class Zodiac:
    def __init__(self, config):
        self.birth_data = config['filepath']['birth_data']
        self.stars_data = config['filepath']['stars_data']

    # Get zodiac name for birthday
    def get_zodiac_name(self, month, day):
        birth_data = from_json(self.birth_data)
        month,day = int(month),int(day)
        for zodiac_name, date in birth_data.items():
            start_date, end_date = date[0], date[1]
            if (month == start_date[0] and day >= start_date[1]) or (month == end_date[0] and day <= end_date[1]):
                return zodiac_name

    # Get centre coordinates of zodiac by using major stars, which consists the zodiac
    def get_zodiac_centre(self, zodiac_name):
        stars_data = from_json(self.stars_data)   
        for zodiac, stars in stars_data.items():
            if zodiac == zodiac_name:
                stars.pop('Graph') 
                centre_ra = np.mean([coords[0] for coords in stars.values()])  # mean of RA
                centre_dec = np.mean([coords[1] for coords in stars.values()])  # mean of Dec
        return centre_ra, centre_dec

    # Compare brightest_pixel instead of the value of flux since it`s impossible to get flux of most of stars from SDSS
    @staticmethod #the method in the class does not depend on the Class instance state,
    def get_brightest_star(stars_coords):
        brightest_star = None
        brightest_pixel = 0
        stars_coords_copy = stars_coords.copy()
        stars_coords_copy.pop('Graph')
        try:
            for star_name, (ra, dec) in stars_coords_copy.items():
                coord = coords.SkyCoord(ra=ra*u.deg, dec=dec*u.deg, frame='icrs')
                res = SkyView.get_images(position=coord, survey=['DSS'], pixels=50)
                                                            
                if res is not None:
                    img_data = res[0][0].data
                    pixel = img_data.max()  
                    if  brightest_pixel < pixel:
                        brightest_pixel = pixel
                        brightest_star = star_name

            return brightest_star
        except Exception as e:
            print(f"An error occurred: {e}")
            return None
```

**`class ImagePlot`**
  
- `get_node_color()` : The name of the brightest star from the get_brightest_star() function gets color in red. 
<br>

- `set_plot()`:  Considering the size of the zodiac`s coordinates range, control the size of ImagePlot with ImagePlot.set_lot().
<br>

- `draw_node()` : Mark the node to visit during BFS traversal.
<br>

- `draw_edge()` : Draw a line on the node linked with the start node. during BFS traversal.
<br>

- `bfs_search_drawing()`  : ImagePlot.bfs_search_drawing() is performed to draw the stars\`s coordinates and line. It is based on the graph information in 'zodiac_stars.json'. In this process, the BFS algorithm is used. Whenever BFS visits each node, they check whether it is the brightest one.
<br>

- `get_image()` : Get the image by performing an HTTP GET request to get data from the SDSS URL.


```python
class ImagePlot:
    def __init__(self, zodiac_name, config):
        self.zodiac_name = zodiac_name
        self.im_size = 1 * u.arcmin
        self.im_pixels = 1024
        self.config = config

    # Return the color of the node based on whether it is the brightest star
    def get_node_color(self, name, brightest_star):
        if name == brightest_star:
            node_colors = 'r'
        else:
            node_colors = 'g'
        return node_colors
    
    # Set up the plot to fit ra, dec values
    def set_plot(self, img_array,  ra_lst, dec_lst, brightest_star):
        ra_min = float(min(ra_lst)) - (self.im_size.to(u.arcsec).value/8) 
        ra_max = float(max(ra_lst)) + (self.im_size.to(u.arcsec).value/8) 
        dec_min = float(min(dec_lst)) - (self.im_size.to(u.arcsec).value/8) 
        dec_max = float(max(dec_lst)) + (self.im_size.to(u.arcsec).value/8) 
        
        plt.title(f'Shape of {self.zodiac_name}')
        plt.xlabel('Right Ascension (RA)')
        plt.ylabel('Declination (DEC)')

        legend_labels = [f'Brightest Star: {brightest_star}']
        legend_elements = [plt.Line2D([0], [0], marker='o', color='b', markerfacecolor='r', markersize=6)]
        plt.legend(legend_elements, legend_labels, loc='upper right', fontsize=7)
        plt.imshow(img_array, extent=[ra_min, ra_max, dec_min, dec_max],aspect='auto')

    # Draw a node on the Plot
    def draw_node(self, name, ra, dec, color):
        plt.scatter(ra, dec, c=color, s=80, zorder=2)
        plt.text(ra, dec, name, ha='center', va='bottom', fontweight='bold', color='white', fontsize=7, zorder=3)

    # Draw an edge on the Plot
    def draw_edge(self, ra1, dec1, ra2, dec2):
        plt.plot([ra1, ra2], [dec1, dec2], linestyle='-', color='b', zorder=1)

    # Draw the graph with BFS algorithm
    def bfs_search_drawing(self, stars_coords, start_node):
        visited = set()
        ra_lst, dec_lst = set(), set()
        queue = deque([start_node])
        graph = stars_coords['Graph']      
        pos = list(stars_coords.values())
        brightest_star = Zodiac.get_brightest_star(stars_coords)
        
        while queue:
            current_node = queue.popleft()
            if current_node not in visited:
                visited.add(current_node)
                next_nodes = graph[str(current_node)]  
                ra1, dec1 = pos[current_node]

                # Drawing node
                name = list(stars_coords.keys())[current_node]
                color = self.get_node_color(name, brightest_star)
                self.draw_node(name, ra1, dec1, color)
                ra_lst.add(ra1)
                dec_lst.add(dec1)

                # Drawing edge
                for next_node in next_nodes:
                    ra2, dec2 = pos[next_node]
                    self.draw_edge(ra1, dec1, ra2, dec2)

                    if next_node not in visited:
                        queue.append(next_node)
                        
            # In the case of node is unconnected 
            if not queue:
                for node in graph.keys():
                    if int(node) not in visited:
                        queue.append(int(node))
                        break
        return ra_lst, dec_lst, brightest_star, visited

    # Get image from SDSS
    def get_image(self,centre_ra, centre_dec, stars_coords):
        base_url = self.config['url']['base_url']
        query_string = urlencode(dict(ra=centre_ra, dec=centre_dec, width=self.im_pixels, height=self.im_pixels,
                                      scale=self.im_size.to(u.arcsec).value / self.im_pixels))
        url = base_url + '?' + query_string

        try:
            response = requests.get(url, stream=True)
            response.raw.decode_content = True
            img = Image.open(BytesIO(response.content))
            img_array = np.array(img)
        except Exception as e:
            print(f"An error occurred: {e}")
            return

        ra_lst, dec_lst, brightest_star, _ = self.bfs_search_drawing(stars_coords, 0)
        self.set_plot(img_array, ra_lst, dec_lst, brightest_star)
```

**`main`** 

 - Perform the overall processing process

```python
def main():
    config = configparser.ConfigParser()   
    config.read('./config.conf', encoding='utf-8') 
    stars_data = config['filepath']['stars_data']
    
    while True:
        try:          
            birthday = input("Please input your Birthday (YYYY-MM-DD): ")
            if not birthday:
                continue
            datetime.strptime(birthday,'%Y-%m-%d')
        except:
            print('Wrong date format')
            continue
        break
    
    year, month, day = birthday.split('-')
    zodiac = Zodiac(config)    
        
    try:
        zodiac_name = zodiac.get_zodiac_name(month, day)
        stars_coords = from_json(stars_data)[zodiac_name]   
        centre_ra, centre_dec = zodiac.get_zodiac_centre(zodiac_name)            
        img = ImagePlot(zodiac_name,config)       
        img.get_image(centre_ra,centre_dec,stars_coords)

    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
